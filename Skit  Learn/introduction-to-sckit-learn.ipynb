## 0. An end-to-end Scikit-Learn workflow

```python
import warnings
warnings.filterwarnings("ignore")
# warnings.filterwarnings("default")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
heart_disease = pd.read_csv("data/heart-disease.csv")
heart_disease
```

```python
# Create a X (features matrix)
x = heart_disease.drop("target",axis=1)

# Create Y (labels)
y = heart_disease["target"]
```

```python
x
```

```python
y
```

```python
# 2. Choose the right model and hyperparameters

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# We will keep the default hyperparameters
clf.get_params()
```

```python
# 3. Fit the model to the trainig data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_test
```

```python
clf.fit(x_train,y_train)
```

```python
# Make a prediction
y_label = clf.predict(np.array([0,2,3,4]))
```

```python
y_preds = clf.predict(x_test)
y_preds
```

```python
x_test
```

```python
y_test

```

```python
x_train
```

```python
# 4. Evaluate the Model on the training and test data
clf.score(x_train,y_train)

```

```python
clf.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test,y_preds)
```

```python
accuracy_score(y_test,y_preds)
```

```python
# Improve a model
# Try differnet amount of n_estimators

np.random.seed(42)
for i in range(10,100,10):
    print(f"Trying model with {i} estimators....")
    clf = RandomForestClassifier(n_estimators=i).fit(x_train,y_train)
    print(f"Model accuracy on test set: {clf.score(x_test,y_test) * 100:.2f}%")
    print("")
```

```python
# Save the Model and load
import pickle 

pickle.dump(clf,open("random_forest_model1.pkl","wb"))
```

```python
loadel_model = pickle.load(open("random_forest_model1.pkl","rb"))
loadel_model.score(x_test,y_test)
```
